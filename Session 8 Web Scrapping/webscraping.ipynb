```python
url = "https://www.opindia.com/latest-news/?nocache"
data = requests.get(url)
data
```

```python
bs = BeautifulSoup(data.content)
bs
```

```python
headings = bs.find_all('h3',class_="entry_title td_module_title")
headings
```
