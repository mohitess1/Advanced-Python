#### Create multiple inputs in function
##### use of Arguments in functions *args

```python
def arguments(*args):
    for i in args:
        print (i)
```

```python
arguments(12,25,36,12,4,2,369,12,1)
```

```python
def add_many(*args):
    """
    this function adds multiple numbers
    """
    #initialise sum to 0
    s = 0
    # apply for loop on args
    for i in args:
        s = s + i
    return s
```

```python
add_many(12,2,0,0,5)
```

```python
12+10+2+3+6
```

```python
s1 = 19+6+9+5+7+1+69
s1
```

#### Getting product of multiple numbers

```python
def product_many(*args):
     #initialise product with 1
    p = 1
    """
    this function multiply multiple numbers
    """
     # apply for loop on args
    for i in args:
        p = p*i
    return p
```

```python
product_many(12,1,2,2)
```

```python
6*10*12*14*11
```

#### Function to calculate Hypotenous of right angle triangle
$ c^2 = a^2 +b^2 $

$ c = sqrt(a^2 + b^2)$

```python
def hypotenous(a,b):
    c = (a**2 + b**2)**(1/2)
    return c
```

```python
h1 = hypotenous(3,4)
h1
```

```python
h2 = hypotenous(12,4)
h2
```

#### Checking Triangle Conditions
sum(two sides)>third remaining side


```python
def is_triangle(X,Y,Z):
    """
    the function tells wheather given 3 sides makes triangle or not
    """
    if (X+Y)>Z and (Y+Z)>X and (Z+X)>Y:
        s = True
    else: 
        s =False
    return s
```

```python
is_triangle(1,2,3)
```

```python
is_triangle(11,12,3)
```

#### Calculating area of triangle with given sides a, b, c
Herons formula s = (a+b+c)/2

area = sqrt(s*(s-a)*(s-b)*(s-c))


```python
def area_triangle(a, b, c):
    if is_triangle(a,b,c)==True:
        s = (a+b+c)/2
        ar = (s*(s-a)*(s-b)*(s-c))**(1/2)
    else:
        ar = None
        print(f'The given sides {a}, {b} and {c} DO NOT Form a triangle.')
    return ar
```

```python
t1 = area_triangle(12, 13, 14)
```
